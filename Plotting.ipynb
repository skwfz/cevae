## Estimated c_yt of the linear model

```python
#Load the results in to convenient numpy arrays
model_byt = np.zeros((len(datasizes), datasize_times))
model_byz2 = np.zeros((len(datasizes), datasize_times))
model_sy = np.zeros((len(datasizes), datasize_times))
for i in range(len(datasizes)):
    for j in range(datasize_times):
        model_byt[i,j] = models_linear_linear[datasizes[i]][j].decoder.y_nn.weight[0,1].detach().numpy()
        model_byz2[i,j] = models_linear_linear[datasizes[i]][j].decoder.y_nn.weight[0,0].detach().numpy()**2
        model_sy[i,j] = np.exp(models_linear_linear[datasizes[i]][j].decoder.y_log_std.detach().numpy())
```

```python
"""Plot the estimated c_yt structural coefficient of the completely linear model. Maybe also the other parameters?"""
fig, ax = plt.subplots(1,3, figsize=(12,4))
ax[0].plot([datasizes[0], datasizes[-1]], 2*[c_yt])
ax[0].set_title("$\\beta_{yt}$")
ax[0].set_xlabel("Sample size")
ax[0].boxplot(model_byt.T, positions=datasizes, widths=400)
ax[0].legend(["$c_{yt}$", "VAE estimate"])
ax[0].tick_params(axis='x',direction='out', length=4, width=1,rotation=45,bottom=True)
ax[0].tick_params(axis='y',direction='out', length=4, width=1,left=True)

ax[1].plot([datasizes[0], datasizes[-1]], 2*[c_yz**2])
ax[1].set_title("$\\beta_{yz}^2$")
ax[1].set_xlabel("Sample size")
ax[1].boxplot(model_byz2.T, positions=datasizes, widths=400)
ax[1].legend(["$c_{yz}^2$", "VAE estimate"])
ax[1].tick_params(axis='x',direction='out', length=4, width=1,rotation=45,bottom=True)
ax[1].tick_params(axis='y',direction='out', length=4, width=1,left=True)

ax[2].plot([datasizes[0], datasizes[-1]], 2*[s_y])
ax[2].set_title("$s_y$")
ax[2].set_xlabel("Sample size")
ax[2].boxplot(model_sy.T, positions=datasizes, widths=400)
ax[2].legend(["$\\sigma_y$", "VAE estimate"])
ax[2].tick_params(axis='x',direction='out', length=4, width=1,rotation=45,bottom=True)
ax[2].tick_params(axis='y',direction='out', length=4, width=1,left=True)

plt.tight_layout()

plt.savefig("Figures/linear_structural_coefs.png")
plt.show()
```

## Causal distance results from all

```python
#Getting the causal distances
model_order = [models_linear_linear, models_linear_pyxt,models_linear_qz,
               models_linear_z2, models_linear_qz_z2, models_linear_pyxt_qz,
               models_linear_everything,models_linear_1_proxy]
avg_causal_dists = len(model_order)*[0]
py_dot_models = len(model_order)*[0]
n = 50
for i in range(len(model_order)):
    avg_causal_dists[i] = np.zeros((len(datasizes), datasize_times))
    py_dot_models[i] = np.zeros((len(datasizes), datasize_times, n, n))
    for k in range(len(datasizes)):
        for l in range(datasize_times):
            model = model_order[i][datasizes[k]][l]
            avg_causal_dist, py_dot_model, py_dot_true, y_range, t_range, pt_true = avg_causal_L1_dist_general(model=model,
                                                c_yt=c_yt, c_yz=c_yz, s_y=s_y, c_t=c_t, s_t=s_t, c_x=c_x, n=n)
            avg_causal_dists[i][k,l] = avg_causal_dist
            py_dot_models[i][k,l,:,:] = py_dot_model
    print("model {} done".format(i))
```

## Explanation for why the most of the time z_dim=2 works

```python
models = [models_linear_z2[5000][0],models_linear_qz_z2[5000][0],models_linear_everything[5000][0]]
fig, ax = plt.subplots(len(models),4,figsize=(16,4.5*len(models)),sharex=True,sharey=True)
for i in range(len(models)):
    model = models[i]
    z = torch.randn((2000,2))
    t = 0*torch.ones((2000,1))
    
    x_pred, t_pred, y_pred, x_std, t_std, y_std = model.decoder(z,t)
    y_mean = y_pred[:,0].detach().numpy().squeeze()
    x_mean1 = x_pred[:,0].detach().numpy().squeeze()
    x_mean2 = x_pred[:,1].detach().numpy().squeeze()
    t_mean = t_pred[:,0].detach().numpy().squeeze()
    ax[i,0].scatter(z[:,0],z[:,1],c=y_mean,vmin=-6,vmax=6)
    ax[i,1].scatter(z[:,0],z[:,1],c=t_mean,vmin=-6,vmax=6)
    ax[i,2].scatter(z[:,0],z[:,1],c=x_mean1,vmin=-6,vmax=6)
    pcm = ax[i,3].scatter(z[:,0],z[:,1],c=x_mean2,vmin=-6,vmax=6)
    #Change settings that are the same for all
    titles = ["$y$ mean","$t$ mean","$x_1$ mean","$x_2$ mean"]
    for j in range(4):
        ax[i,j].tick_params(axis='both',direction='out', length=4, width=1,left=True,bottom=True)
        ax[i,j].set_title(titles[j])
        ax[i,j].set_xlabel("$z_1$")
        ax[i,j].set_ylabel("$z_2$")
    #Plot the arrows
    if not model.decoder.p_y_zt_nn:
        wy = model.decoder.y_nn.weight[0,:-1].detach().numpy()
        ax[i,0].arrow(0,0,wy[0],wy[1],width=0.1,length_includes_head=True,facecolor="w")
    if not model.decoder.p_t_z_nn:
        wt = model.decoder.t_nn.weight[0,:].detach().numpy()
        ax[i,1].arrow(0,0,wt[0],wt[1],width=0.1,length_includes_head=True,facecolor="w")
    if not model.decoder.p_x_z_nn:
        wx1 = model.decoder.x_nns[0].weight[0,:].detach().numpy()
        ax[i,2].arrow(0,0,wx1[0],wx1[1],width=0.1,length_includes_head=True,facecolor="w")
        wx2 = model.decoder.x_nns[1].weight[0,:].detach().numpy()
        ax[i,3].arrow(0,0,wx2[0],wx2[1],width=0.1,length_includes_head=True,facecolor="w")
plt.tight_layout()
fig.subplots_adjust(right=0.96)
cbar_ax = fig.add_axes([0.97, 0.37, 0.01, 0.28])
fig.colorbar(pcm, cax=cbar_ax)
plt.savefig("Figures/correct_latent_use.png")
plt.show()
```

## ATE and posterior approximation

```python
#Getting the ATEs
model_order = [models_bl_basic, models_bl_strong_nonlinear,models_bl_zdim2,
               models_bl_decoder_nn, models_bl_zdim2_decoder_nn]
model_ATEs = len(model_order)*[0]
for i in range(len(model_order)):
    model_ATEs[i] = np.zeros((len(datasizes), datasize_times))
    for k in range(len(datasizes)):
        for l in range(datasize_times):
            model = model_order[i][datasizes[k]][l]
            model_ate = linear_binary_ty_ate_3(model, model.z_dim)
            model_ATEs[i][k,l] = model_ate
    print("model {} done".format(i))
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

model_titles = ["Basic model", "Basic model, strong nonlinearity"]
fig, ax = plt.subplots(2,2, figsize=(8,8), sharey=False)
models_considered = [0,1]
for i in range(len(models_considered)):
    index = models_considered[i]
    ax[i][0].plot([datasizes[0],datasizes[-1]], 2*[true_ate])
    ax[i][0].boxplot(model_ATEs[i].T, positions=datasizes, widths=400)
    ax[i][0].legend(["True ATE"])
    
    t = torch.Tensor([[0]])
    y = torch.Tensor([[1]])
    x = torch.Tensor([[0,0]])
    z_mean, z_std = model_order[index][5000][0].encoder(x,t,y)
    z_mean = z_mean.detach().numpy().squeeze()
    z_std = z_std.detach().numpy().squeeze()
    lim = 2
    n = 300
    z = np.linspace(-lim,lim,n)
    #NOTE: this is manually set here, doesn't change if you change it elsewhere
    t_a_ = 1 if index==0 else 50
    pz_xty = scipy.stats.norm.pdf(x0, loc=c_x[0]*z, scale=s_x[0]) \
        *scipy.stats.norm.pdf(x1, loc=c_x[1]*z, scale=s_x[1]) \
        *(1-sigmoid(t_a_*z + t_b))*sigmoid(y_a0*z + y_b0) \
        *scipy.stats.norm.pdf(z)
    pz_xty = pz_xty / (pz_xty.sum()*2*lim/n)
    ax[i,1].plot(z, scipy.stats.norm.pdf(z,loc=z_mean, scale=z_std))
    ax[i,1].plot(z,pz_xty)
    ax[i,1].legend(["Variational approximation", "True posterior"])
    #for j in range(len(datasizes)):
    #    xpositions = np.array(datasize_times*[datasizes[j]]) + 50*(np.random.rand(datasize_times)-0.5)
    #    ypositions = model_ATEs[i][j,:] #+ 0.001*np.random.randn(datasize_times)
    #    ax[i][0].plot(xpositions,ypositions,'o',color='C0',markersize=4)
    
    #ax[i][0].set_title(model_titles[i])
    ax[i][0].tick_params(axis='x',direction='out', length=4, width=1,rotation=45,bottom=True)
    ax[i][0].tick_params(axis='y',direction='out', length=4, width=1,left=True)
    ax[i][1].tick_params(axis='x',direction='out', length=4, width=1,rotation=0,bottom=True)
    ax[i][1].tick_params(axis='y',direction='out', length=4, width=1,left=True)
plt.tight_layout()
plt.savefig("Figures/ATEs_posteriors.png")
plt.show()
```
